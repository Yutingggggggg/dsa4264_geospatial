```python
# Apply the filter for DAY_TYPE as WEEKDAY and the specified TIME_PER_HOUR ranges
filtered_passenger_volume_bus = filtered_passenger_volume_bus[
    (filtered_passenger_volume_bus['DAY_TYPE'] == 'WEEKDAY') & 
    (
        (filtered_passenger_volume_bus['TIME_PER_HOUR'] >= 7) & (filtered_passenger_volume_bus['TIME_PER_HOUR'] < 11) |
        (filtered_passenger_volume_bus['TIME_PER_HOUR'] >= 17) & (filtered_passenger_volume_bus['TIME_PER_HOUR'] < 21)
    )
]

# Sort the filtered DataFrame by TIME_PER_HOUR in ascending order
filtered_passenger_volume_bus = filtered_passenger_volume_bus.sort_values(by='TIME_PER_HOUR', ascending=True)
filtered_passenger_volume_bus
```

```python
filtered_passenger_volume_bus = filtered_passenger_volume_bus[filtered_passenger_volume_bus['DAY_TYPE'] == 'WEEKDAY']

# Separate the morning peak hours (7 to 10) and evening rush hours (17 to 20)
morning_peak = filtered_passenger_volume_bus[
    (filtered_passenger_volume_bus['TIME_PER_HOUR'] >= 7) & (filtered_passenger_volume_bus['TIME_PER_HOUR'] < 11)
]

evening_rush = filtered_passenger_volume_bus[
    (filtered_passenger_volume_bus['TIME_PER_HOUR'] >= 17) & (filtered_passenger_volume_bus['TIME_PER_HOUR'] < 21)
]

# Calculate total tap-in and tap-out volumes for each bus stop during morning peak hours
morning_peak_totals = morning_peak.groupby('BusStopCode')[['TOTAL_TAP_IN_VOLUME', 'TOTAL_TAP_OUT_VOLUME']].sum().reset_index()

# Calculate total tap-in and tap-out volumes for each bus stop during evening rush hours
evening_rush_totals = evening_rush.groupby('BusStopCode')[['TOTAL_TAP_IN_VOLUME', 'TOTAL_TAP_OUT_VOLUME']].sum().reset_index()

# Merge the two results for comparison
totals_combined = pd.merge(morning_peak_totals, evening_rush_totals, on='BusStopCode', how='outer', suffixes=('_Morning', '_Evening'))

totals_combined['Total_Tap_In'] = totals_combined['TOTAL_TAP_IN_VOLUME_Morning'] + totals_combined['TOTAL_TAP_IN_VOLUME_Evening']
totals_combined['Total_Tap_Out'] = totals_combined['TOTAL_TAP_OUT_VOLUME_Morning'] + totals_combined['TOTAL_TAP_OUT_VOLUME_Evening']
totals_combined['Average_Tap_In'] = (totals_combined['TOTAL_TAP_IN_VOLUME_Morning'] + totals_combined['TOTAL_TAP_IN_VOLUME_Evening'])/8
totals_combined['Average_Tap_Out'] = (totals_combined['TOTAL_TAP_OUT_VOLUME_Morning'] + totals_combined['TOTAL_TAP_OUT_VOLUME_Evening'])/8
totals_combined['Total_Tap_In'] = totals_combined['TOTAL_TAP_IN_VOLUME_Morning'] + totals_combined['TOTAL_TAP_IN_VOLUME_Evening']
totals_combined['Total_Tap_Out'] = totals_combined['TOTAL_TAP_OUT_VOLUME_Morning'] + totals_combined['TOTAL_TAP_OUT_VOLUME_Evening']
totals_combined['Average_Tap'] = (totals_combined['Total_Tap_In'] + totals_combined['Total_Tap_Out']) / 8

# Display the results
totals_combined
```

```python
# Sort the totals_combined DataFrame by Average_Tap in descending order
totals_combined = totals_combined.sort_values(by='Average_Tap', ascending=False)
# Display the resulting DataFrame
totals_combined
```

```python
totals_combined_coordinate = totals_combined.merge(
    bus_stops_coordinates,
    left_on='BusStopCode',   # Ensure this matches the column name exactly
    right_on='BusStopCode',  # Ensure this matches the column name exactly
    how="left"
)

# Assuming 'BusStopCode' is the common column in both DataFrames
totals_combined_coordinate = totals_combined_coordinate.merge(
    bus_route_names[['BusStopCode', 'Description']],  # Include BusStopCode to maintain the relationship
    on='BusStopCode',  # Specify the column to merge on
    how='left'         # Use left join to keep all records from totals_combined_coordinate
)

totals_combined_coordinate = totals_combined_coordinate.drop_duplicates()
totals_combined_coordinate
```

```python
import folium
from folium import PolyLine

# Calculate the min and max latitude and longitude to fit all points on the map
min_lat, max_lat = totals_combined_coordinate['Latitude'].min(), totals_combined_coordinate['Latitude'].max()
min_lon, max_lon = totals_combined_coordinate['Longitude'].min(), totals_combined_coordinate['Longitude'].max()

# Initialize the map centered around the data and set a broad initial zoom level
bus67_bus_stop = folium.Map(location=[(min_lat + max_lat) / 2, (min_lon + max_lon) / 2], zoom_start=10)
bus67_bus_stop.fit_bounds([[min_lat, min_lon], [max_lat, max_lon]])

# Define a color function based on Average_Tap threshold
def get_marker_color(avg_tap):
    return 'red' if avg_tap > totals_combined_coordinate['Average_Tap'].mean() else 'blue'

# Collect coordinates for route lines
route_coordinates = []

# Plot each bus stop as an individual marker, no clustering
for _, row in totals_combined_coordinate.iterrows():
    route_coordinates.append((row['Latitude'], row['Longitude']))

    # Add each bus stop as a circle marker that stays visible at all zoom levels
    folium.CircleMarker(
        location=(row['Latitude'], row['Longitude']),
        radius=5,
        color=get_marker_color(row['Average_Tap']),
        fill=True,
        fill_color=get_marker_color(row['Average_Tap']),
        fill_opacity=0.7,
        tooltip=f"Bus Stop: {row['BusStopCode']}<br>Average Tap: {row['Average_Tap']}<br>Bust Stop Name: {row['Description']}"
    ).add_to(bus67_bus_stop)


# Display the map
bus67_bus_stop
```

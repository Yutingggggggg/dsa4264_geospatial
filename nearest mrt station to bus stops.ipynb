```python
import geopandas as gpd
import pandas as pd
from tqdm import tqdm
from shapely.geometry import Point

def calculate_distance_to_mrt(bus_volume_coordinates, train_station_exits_gdf):
    # Bus origin points
    bus_origin_gdf = gpd.GeoDataFrame(bus_volume_coordinates, 
                                      geometry=gpd.points_from_xy(bus_volume_coordinates['Origin Longitude'], bus_volume_coordinates['Origin Latitude']),
                                      crs="EPSG:4326")
    bus_origin_gdf_3857 = bus_origin_gdf.to_crs(3857)
    train_station_exits_gdf_3857 = train_station_exits_gdf.to_crs(3857)
    
    score = []
    for i, row in tqdm(bus_origin_gdf_3857.iterrows(), total=bus_origin_gdf_3857.shape[0]):
        distances = train_station_exits_gdf_3857.distance(row['geometry'])
        min_distance = float(distances.min())
        nearest_station = train_station_exits_gdf_3857.loc[distances.idxmin(), 'stn_name']
        
        score.append({
            'ORIGIN_PT_CODE': row['ORIGIN_PT_CODE'],
            'Origin Latitude': row['Origin Latitude'],
            'Origin Longitude': row['Origin Longitude'],
            'Nearest MRT Station': nearest_station,
            'Distance to Nearest MRT': min_distance
        })
    
    return pd.DataFrame(score)

# Closer = higher score
def normalize(df, distance_col):
    max_distance = df[distance_col].max()
    df['Normalized Distance'] = 1 - (df[distance_col] / max_distance)  
    return df

# Incorporate Passenger volume (TOTAL_TRIPS)
def compute_weighted_overlap(bus_volume_coordinates, overlap_df, volume_col='TOTAL_TRIPS'):
    bus_overlap = pd.merge(bus_volume_coordinates, overlap_df, on='ORIGIN_PT_CODE')
    
    # Formula for weighted score
    bus_overlap['Weighted Overlap'] = bus_overlap['Normalized Distance'] * bus_overlap[volume_col]
    route_overlap_scores = bus_overlap.groupby(['ORIGIN_PT_CODE']).agg(
        total_overlap=pd.NamedAgg(column='Weighted Overlap', aggfunc='sum'),
        avg_overlap=pd.NamedAgg(column='Weighted Overlap', aggfunc='mean'),
        total_volume=pd.NamedAgg(column=volume_col, aggfunc='sum')
    ).reset_index()

    return route_overlap_scores

overlap_df = calculate_distance_to_mrt(bus_volume_coordinates, train_station_exits_gdf)
normalized_overlap_df = normalize(overlap_df, 'Distance to Nearest MRT')
route_overlap_scores = compute_weighted_overlap(bus_volume_coordinates, normalized_overlap_df)
sorted_routes = route_overlap_scores.sort_values(by='total_overlap', ascending=False)

```
